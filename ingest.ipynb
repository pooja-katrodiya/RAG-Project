```python
#Read the pdf
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path: str) -> str:
    doc = fitz.open(pdf_path)
    full_text = []
    for page_num, page in enumerate(doc, start=1):
        text = page.get_text("text")
        if text.strip():
            full_text.append(f"\n[Page {page_num}]\n{text}")
    doc.close()
    return "\n".join(full_text)

# Call it like this:
text = extract_text_from_pdf(r"C:\Users\kruta\Desktop\ML PREP\RAG\data\paperNN.pdf")
print(text[:500])  # Preview first 500 chars
```

```python
#split the paper into chunks
def chunk_text(text: str, chunk_size: int = 500, overlap: int = 100) -> list[str]:
    words = text.split()
    chunks = []

    start = 0
    while start < len(words):
        end = start + chunk_size
        chunk = " ".join(words[start:end])
        chunks.append(chunk)
        start += chunk_size - overlap

    return chunks
```

```python
pdf_path = r"C:\Users\kruta\Desktop\ML PREP\RAG\data\paperNN.pdf"
text = extract_text_from_pdf(pdf_path)
chunks = chunk_text(text)

print("Total characters:", len(text))
print("Total chunks:", len(chunks))
print("\nchunk:\n")
print(chunks[5])
```
