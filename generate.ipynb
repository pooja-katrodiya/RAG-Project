```python
# Generate an answer with Ollama
import requests


def generate_answer(question: str, retrieved_chunks: list[tuple[str, str]], model_name: str = "llama3.2") -> str:
    context = "\n\n".join(
        [f"{chunk_id}: {chunk_text}" for chunk_id, chunk_text in retrieved_chunks]
    )

    prompt = f"""
You are a research paper assistant.

Answer the user's question using only the context below.
If the answer is not in the context, say: "I could not find the answer in the provided paper."

Context:
{context}

Question:
{question}

Also mention which chunk IDs support the answer.
"""

    response = requests.post(
        "http://localhost:11434/api/generate",
        json={
            "model": model_name,
            "prompt": prompt,
            "stream": False
        },
        timeout=120
    )

    response.raise_for_status()
    data = response.json()
    return data["response"]
```
